```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

import warnings

# Suppress all warnings
warnings.filterwarnings('ignore')
```

```python
# Load in the full dataset
df = pd.read_csv("encoded_data.csv")

# Make a copy of the full dataset; this is the one we will manipulate while still retaining the full original data
df_subset = df
Just_AI_Satisfaction = df_subset['AI_Satisfaction']
df_subset = df_subset.drop('AI_Satisfaction', axis = 1) # taking out target variable
df_subset.head()
```

```python
# Scale the data like within LDA
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df_subset)

```

```python
# using the elbow method to determine optimal number of clusters
SSE = []  # Sum of Squared Errors

# Looking at clusters 1 - 10
for i in range(1, 11):  
    kmeans = KMeans(n_clusters = i, random_state = 3)
    kmeans.fit(scaled_df)
    SSE.append(kmeans.inertia_)

# Plot the Elbow method
plt.plot(range(1, 11), SSE)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE (sum of squared errors)')
plt.show()
```

```python
# Apply K-means from the sklearn package 
kmeans = KMeans(n_clusters = 6, random_state = 3)  
df_subset['Cluster'] = kmeans.fit_predict(scaled_df)
```

```python
# Get the two best clusters from the k-means
cluster_sizes = df_subset['Cluster'].value_counts()
top_clusters = cluster_sizes.nlargest(2).index

```

```python
# add them back into the data set and put the arget variable at the end
df_subset['KmeansCluster1'] = np.where(df_subset['Cluster'] == top_clusters[0], 1, 0)
df_subset['KmeansCluster2'] = np.where(df_subset['Cluster'] == top_clusters[1], 1, 0)

df_subset['AI_Satisfaction'] = Just_AI_Satisfaction
print(df_subset.head())
```

```python
from os import read
# Export DataFrame to CSV
df_subset.to_csv('FE_final_data.csv', index=False)

```
